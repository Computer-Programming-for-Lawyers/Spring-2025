<h3> Hot Lanes in five sentences.</h3>
<h4>
1. If it is morning rush-hour, then the HOT-lanes are in effect east-bound. <br><br>
2. This means that a commuter can only take I-42 east-bound if they have three or more people in the car or they're willing to pay the tolls. <br><br>
3. If it's evening rush-hour, then the HOT-lanes are in effect west-bound. <br><br>
4. This means that a commuter can only take I-42 west-bound if they have three or more people in the car or they're willing to pay the tolls. <br><br>
5. At any other time of day or in any other direction, the HOT-lanes aren't in effect, so the commuter is free to use I-42 with no worries. <br><br>
</h4>

```python
### Truth Tables ###

# time      | direction | HOT_lanes_in_effect
# -------------------------------------------
# no        | east      |  False
# no        | west      |  False
# morning   | east      |  True
# morning   | west      |  False
# evening   | east      |  False
# evening   | west      |  True
```

```python
# Putting the truth table into practice (code)
# assume inputs are already given.
time = input("What time is it (morning/evening/no):")
direction = input("east or west?:")
pax = input("how many people riding?:")
# Eastbound hotlane restrictions in effect

eastbound_hotlane_in_effect = (time == 'morning' and direction == 'east')
# Westbound hotlane restrictions in effect
westbound_hotlane_in_effect = (time == 'afternoon' and direction == 'west')

# Number of people.
under_floor = (pax == '1' or pax == '2')

# True at all times/direction/pax EXCEPT (east & morning OR west & afternoon) AND pax < 3
allowed_in_hotlanes = not((westbound_hotlane_in_effect or eastbound_hotlane_in_effect) and under_floor)

print(allowed_in_hotlanes)
```

<h3> Using If/Elif/Else Logic and While Loops: Meet Eliza </h3>

http://psych.fullerton.edu/mbirnbaum/psych101/Eliza.htm

Let's build a much less sophisticated replica? of Eliza.

```python
# Eliza.

# Eliza introduces itself and invites user input.
print()
print ("Hello, I am Eliza. What's on your mind?")
i = 0


# How Eliza should work: (what tools should we use?)
# 1. Eliza should stay engaged until told to quit. What should we use here?
while True:
    
    #add 1 to the counter as we enter the user input phase.
    i += 1
    
    # Take input from user.
    command = input("> ")
    # You can also print how many commands have been entered like this:
    # command = input(str(i) + "> ")
    # print(command)

# 2. Eliza should give tailored responses to certain command keywords. What should we use here?
    if command == 'Hello':
        print("Hi, how are you?")
# 3. Commands include: Hello; Weather;
    elif command == 'Weather':
        print("It is sunny today.") 
# 4 Nested Loops: Amazon Alexa Monopoly mode
    elif command == "Alexa":
        # Demonstrating nested control flow
        print()
        print("You have unlocked my secret Jeff Bezos control module. What would you like to do?")
        while True:
            # Note how the hunger module does not activate while inside the nested (Alexa) loop.
            command = input("Alexa> ")
            if command == "Hello":
                print("Hi. Could I interest you in a $99/year prime membership?")
            elif command == "quit" or command == "exit":
                print("Hah! You think you can ever escape our clutches???")
            # Note how the break in the nested loop only breaks the inner-loop (Alexa) but not the outer (Eliza)
            elif command == "Lina Khan" or command == 'antitrust':
                print("Fine. Back to Eliza.")
                break
            else:
                print('You seem confused. Users like you also like the command "Hello".')
# 5. Eliza should say goodbye when session is over.
    elif command == 'feed':
        print('Nom Nom.')
        i = 0
    elif command == 'quit' or command == 'exit':
        break
    elif command == 'Hello':
        # will this code ever be reachable?
        print("from the other side")
# 6. Catchall response to everything else.
    else:
        print("I don't understand.")
        # Do this block second
    
    # Adding Hunger module using modulo. Using a second if-elif-else inside the loop
    # to add more behavior to the loop that is sensitive to user input (command count).
    # Eliza gets hungry
    since_last_meal = i % 3
    if since_last_meal == 1:
        print("I'm peckish.")
    elif since_last_meal == 2:
        print("I'm starving!")
    elif i == 0:
        print("I'm full.")
    else:
        # Will this ever be reached?
        print("That's it, I'm eating!")
    
    print()
    print ("Hello, I am Eliza. What's on your mind?")

print("Good bye!")

```
