# What is Anagram Spell, conceptually?
##### 0. Spell check the list (e.g., Alice.txt)
##### 1. Find silly words e.g., 'badl'
##### 2. See if bad word sorted (e.g., 'abdl') matches the sorted (a-z) version of any correct words ('bald'->'abdl')
##### 3. If match found, print. If not, print '(No Suggestions)'

### In other words, Anagram Spell is:
### If a word doesn't match the dictionary, see if its sorted anagram matches the sorted version of any words in the dictionary.




```python
# Let's Code Anagram Spell
# First import the files
# This will be declaration.txt
input_filename = 'declaration.txt'
declaration_of_independence = open(input_filename).read().split()

# This will be word_list.txt
word_list_filename = 'word_list.txt'
word_list = open(word_list_filename).read().split('\n')

# Then what? Anagrams?
# alphabetically sorted version of a word. E.g., word -> dorw

dictionary_anagrams = []

for i in word_list:
    dictionary_anagrams.append(sorted(i))

# sorted(word_list[34534]) == dictionary_anagrams[34534]

for decl_word in declaration_of_independence:
    # what to do if i is in fact a correct word? -> Do nothing move on. Continue.
    if decl_word in word_list:
        continue
    
    # Okay word is misspelled, let's anagram it.
    sorted_word_from_decl = sorted(decl_word)
    # Look for anagram matches in the dictionary.
    matches = []

    for i in range(len(word_list)): # 0 ~ 235885
        if sorted_word_from_decl == dictionary_anagrams[i]:
            matches.append(word_list[i])
    
    # This method is really slow.
    # for i in word_list:
    #     sorted_word_from_list = sorted(i)
    #     # if match append/add to matches
    #     if sorted_word_from_list == sorted_word_from_decl:
    #         matches.append(i)
    
    # when we get here matches may have 0, 1, or more elements.
    # How len() works with lists and strings
    # len([]) -> 0
    # len(['hello', 'my', 'name']) -> 3
    # len('hello') -> 5

    if len(matches) == 0:
        print(decl_word+' -> '+'(No matches found)')
    else:
        print(decl_word+' -> '+', '.join(sorted(matches)))




```
