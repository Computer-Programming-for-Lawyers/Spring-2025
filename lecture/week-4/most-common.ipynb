### Activity: Given a text file full of prose, find the most common word.
If two or more words are tied for most common, print all of them

```python
# print the number of times each word appears, using a list

filename = 'brandeis.txt' #input("Enter filename: ")

brandeis = open(filename).read().split() # empty split breaks by whitespace

brandeis[:10] #first 10 list elements
```

```python
# attempt one using a list
word_count = []

for word in brandeis:
    word_count.append(brandeis.count(word))

sorted(word_count,reverse=True)
```

```python
# attempt two using a dictionary

word_count = {}

#create lowercase version of brandeis
brandeis_lower = []
for word in brandeis:
    brandeis_lower += [word.lower()]

for word in brandeis_lower:
    # add a value to the dictionary that includes the key (word) and the value (the count of how 
    # often that word appears)
    
    if word not in word_count:
        word_count[word] = brandeis_lower.count(word)

word_count
```

```python
# get the VALUE of the highest count word
top_value = sorted(word_count.values(),reverse=True)[0]

print("top value:",top_value)

# for each key in the dictionary
for item in word_count:

    #if the value associated with that key equals the top value, then print the key
    if word_count[item] == top_value:
        print(f"The word that shows up the most is: {item}.")
        print(f"It shows up {word_count[item]} times.")
```
